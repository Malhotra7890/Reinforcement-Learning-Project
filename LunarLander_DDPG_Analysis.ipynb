LUNAR LANDAR ANALYSIS

```python
import matplotlib.pyplot as plt
import json
import numpy as np

with open("ddpg_training_data.json", "r") as f:
    all_episodes_data = json.load(f)

# Rewards (raw values)
rewards = np.array([ep["reward"] for ep in all_episodes_data])
episodes = np.arange(1, len(rewards) + 1)

plt.figure(figsize=(12,5))
plt.plot(episodes, rewards, color='blue', linewidth=1.8)

# ---------- Y-TICKS AT 200 INTERVAL ----------
interval = 200
y_min = int((min(rewards) // interval) * interval) - interval
y_max = int((max(rewards) // interval) * interval) + interval

plt.yticks(np.arange(y_min, y_max + 1, interval))
plt.ylim(y_min, y_max)

# Labels
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.title("Reward vs Episodes")
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()




```

```python

# Choose an episode to visualize step-wise parameters
episode = all_episodes_data[-1]  # last episode, you can change index
steps = list(range(len(episode["main_thrusts"])))

# Main & Side Thrusts

plt.figure(figsize=(6,4))
plt.plot(steps, episode["main_thrusts"], label="Main Thrust (Vertical)", color='blue')
plt.plot(steps, episode["side_thrusts"], label="Side Thrust (Horizontal)", color='orange')
plt.xlabel("Step")
plt.ylabel("Thrust Value")
plt.title("Main & Side Thrust per Step")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Velocities
plt.figure(figsize=(6,4))
plt.plot(steps, episode["velocity_x"], label="Velocity X", color='green')
plt.plot(steps, episode["velocity_y"], label="Velocity Y", color='red')
plt.xlabel("Step")
plt.ylabel("Velocity")
plt.title("Horizontal & Vertical Velocities per Step")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
